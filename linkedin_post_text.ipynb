```python
def generate_text(query):
    prompt = PromptTemplate(
        input_variables=["topic"],
        template="""
        Write a professional and short LinkedIn post about {topic}.
        Keep it simple, clear, and engaging.
        """
    )

    llm = ChatGoogleGenerativeAI(
        model="models/gemini-2.5-flash",
        api_key=api_key,
        temperature=0.7
    )

    chain = prompt | llm
    response = chain.invoke({"topic": query})

    return response.content
```

```python
def generate_image(prompt, output_path="generated_image.png"):
    client = InferenceClient(
        model="black-forest-labs/FLUX.1-schnell",
        token=HF_API_KEY
    )

    image = client.text_to_image(prompt)
    image.save(output_path)

    return output_path
```

```python
def ai_generate_pipeline(query):
    # 1. Generate text
    text = generate_text(query)

    # 2. Generate image
    image_prompt = f"Professional illustration representing {query}"
    image_path = generate_image(image_prompt)

    return {
        "text": text,
        "image_path": image_path
    }

```

```python
query = "sql interview questions"

result = ai_generate_pipeline(query)

print("Generated Text:\n")
print(result["text"])

print("\nImage saved at:", result["image_path"])
```
